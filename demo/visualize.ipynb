```python
import matplotlib.pyplot as plt
import json
import os
```

```python
outdir = "output"
net = "MCNN"
dataset = "SHHB"
file_name = "record.json"
record_path = os.path.join("../", outdir, net, dataset, file_name)
print("load record data from: %s" % record_path)
with open(record_path, 'r') as f:
    record = json.load(f)
    f.close()
```

```python
# new figure
plt.figure(figsize=(15, 5))
# loss_history
plt.subplot(2,2,1)
plt.title('loss_history')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.plot(record['loss_history'])
plt.grid(axis='x')
plt.grid(axis='y')
# mae_history
plt.subplot(2,2,2)
plt.title('mae_history')
plt.xlabel('epoch')
plt.ylabel('mae')
plt.plot(record['mae_hisrory'])
plt.grid(axis='x')
plt.grid(axis='y')
# mse_history
plt.subplot(2,2,3)
plt.title('mse_history')
plt.xlabel('epoch')
plt.ylabel('mse')
plt.plot(record['mse_history'])
plt.grid(axis='x')
plt.grid(axis='y')
# show
plt.tight_layout()
plt.show()
```
